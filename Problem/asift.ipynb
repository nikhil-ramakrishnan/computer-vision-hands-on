```python
img_1 = "/content/drive/MyDrive/Projects/cv/dataset/uncc-dataset/cereal_box_1/separated/cereal_box_1_1_1_separated.png"
img_2 = "/content/drive/MyDrive/Projects/cv/dataset/uncc-dataset/cereal_box_1/separated/cereal_box_1_2_1_separated.png"
img_3 = "/content/drive/MyDrive/Projects/cv/dataset/uncc-dataset/cereal_box_1/separated/cereal_box_1_3_1_separated.png"
img_4 = "/content/drive/MyDrive/Projects/cv/dataset/uncc-dataset/cereal_box_1/separated/cereal_box_1_4_1_separated.png"
img_5 = "/content/drive/MyDrive/Projects/cv/dataset/uncc-dataset/cereal_box_1/separated/cereal_box_1_5_1_separated.png"
img_6 = "/content/drive/MyDrive/Projects/cv/dataset/uncc-dataset/cereal_box_1/separated/cereal_box_1_6_1_separated.png"


img1 = cv2.imread(img_1, 0)
img2 = cv2.imread(img_2, 0)
img3 = cv2.imread(img_3, 0)
img4 = cv2.imread(img_4, 0)
img5 = cv2.imread(img_5, 0)
img6 = cv2.imread(img_6, 0)


images_pose = [img1, img2, img3, img4, img5, img6]
imagesout_pose = []
h_values_pose = []
kpoints_pose = []
kp1s_pose = []
kp2s_pose = []
num_inliers_pose = []
```

```python
for i in range(0, len(images_views)-1):
      print("Printing image --> ${0}, ${1}".format(i, i+1))
      imgout,h, kpoints, kp1, kp2,n_inliers = find_asift(images_views[i], images_views[i+1])
      imagesout_views.append(imgout)
      h_values_views.append(h)
      kpoints_views.append(kpoints)
      kp1s_views.append(kp1)
      kp2s_views.append(kp2)
      num_inliers_pose.append(n_inliers)
```

```python
for i in range(0, len(images_pose)):
  for j in range(0, len(images_pose)):
    if i != j :
      print("Printing image --> ${0}, ${1}".format(i, j))
      imgout,h, kpoints, kp1, kp2, n_inliers = find_asift(images_pose[i], images_pose[j])
      imagesout_pose.append(imgout)
      h_values_pose.append(h)
      kpoints_pose.append(kpoints)
      kp1s_pose.append(kp1)
      kp2s_pose.append(kp2)
      num_inliers_pose.append(n_inliers)
```
