### Import Packages

```python
from PIL import Image,ImageChops     #To open,manipulate and save different image format
from IPython.display import display  #To display image in the jupyter notebook
from IPython.display import Markdown #To give title to images
```

### Opening the images

```python
img1=Image.open("./dataset/img2 (1).png")
img2=Image.open("./dataset/img2 (2).png")
```

### Resizing the image

```python
new_size = (img1.size[0] // 2, img1.size[1] // 2) # Sets the width and height
res_img1 = img1.resize(new_size)
res_img2=img2.resize(new_size)
# Resizing is done so that it accurately fits the notebook acoording to our need
```

### Display the images Title

```python
display(Markdown("Image1"))
display(res_img1)
display(Markdown("Image2"))
display(res_img2)
```

### Finding the difference between two images

```python
diff=ImageChops.difference(res_img1,res_img2)
display(diff)
```

### No difference is shown hence we set a threshold value so that even the minute differences can be obtained

```python
# Set a threshold value for the pixel difference
threshold = 75

# Create a mask image that shows the pixels with a difference above the threshold
mask = diff.convert('L').point(lambda x: x > threshold and 255)
# diff.convert('L') converts the image to grayscale. This is necessary because the point() method that we will use next operates on grayscale images.
# .point(lambda x: x > threshold and 255) applies a mapping function to each pixel value in the image. The mapping function 
#  takes a single argument x, which represents the pixel value, and returns a new pixel value based on the threshold condition.

# Apply the mask to the difference image to make the differences more visible
diff.putalpha(mask) #Sets up the transparency channel

# Display the resulting difference image
display(diff)
```
